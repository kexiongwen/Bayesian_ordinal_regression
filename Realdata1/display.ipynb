```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
beta1=np.load('beta1.npy')
```

```python
lam1 = np.load('lam1.npy')
```

```python
beta1_mean=np.mean(beta1,axis=1)
```

```python
location1=np.where(np.abs(np.mean(beta1,axis=1)/np.std(beta1,axis=1))>1.96)
```

```python
location2=np.where(np.abs(np.mean(beta1,axis=1)/np.std(beta1,axis=1))>1.64)
```

```python
f,axes=plt.subplots(5,3,figsize=(25,25))
axes=axes.ravel()

for i in range(15):
    sns.kdeplot(data=beta1[location1,:][0,i,:],ax=axes[i])
    axes[i].set_xlabel(location1[0][i])
    axes[i].set_ylabel(r'Density of $\beta$')

plt.show()
```
