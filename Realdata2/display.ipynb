```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
beta2=np.load('beta2.npy')
```

```python
lam2 = np.load('lam2.npy')
```

```python
beta2_mean=np.mean(beta2,axis=1)
```

```python
location1=np.where(np.abs(np.mean(beta2,axis=1)/np.std(beta2,axis=1))>1.96)
```

```python
location2=np.where(np.abs(np.mean(beta2,axis=1)/np.std(beta2,axis=1))>1.64)
```

```python
f,axes=plt.subplots(5,3,figsize=(25,25))
axes=axes.ravel()

for i in range(14):
    sns.kdeplot(data=beta2[location1,:][0,i,:],ax=axes[i])
    axes[i].set_xlabel(location1[0][i])
    axes[i].set_ylabel(r'Density of $\beta$')

plt.show()
```
